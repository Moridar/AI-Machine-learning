```python
# import required libraries
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# importing CarPrice_Assignment.csv with pandas

CarData= pd.read_csv('Car_Price.csv')

CarData.head()
```

```python
# shape of the dataset
CarData.shape
```

```python
# Display general information about the dataset
# This includes the count of non-null entries per column, the data type of each column, and memory usage.
CarData.info()
```

```python
# Generate descriptive statistics for numerical columns
# Provide information like mean, standard deviation, min, max, and quartiles.
CarData.describe()
```

```python
# Extract the car's company's name from the variable 'CarName' into the variable 'CarCompany'

CarData['CarCompany'] = CarData['CarName'].apply(lambda x: x.split(' ')[0])

CarData.head()
```

```python
# #checking the unique values under "CarCompany"
CarData['CarCompany'].unique()
```

```python
# replacing the mis-spelling with correct ones

CarData['CarCompany'] = CarData['CarCompany'].apply(lambda x: 'toyota' if x == 'toyouta' else 'volkswagen' if x == 'vw' or x == 'vokswagen' else 'nissan' if x == 'Nissan' else 'mazda' if x == 'maxda' else 'porsche' if x == 'porcshce' else x)

# again checking the unique values
CarData.CarCompany.unique()
```

```python
# Visualising 'CarComapny' variable aginst dependent variable 'price' via boxplot

plt.figure(figsize=(16, 4.5))
sb.set_style('ticks')
sb.set_context('paper')
boxplot = sb.boxplot(x = 'CarCompany', y= 'price', hue = 'CarCompany', data = CarData, palette=sb.husl_palette(22, s=0.90),
           flierprops=dict(marker='d', markersize=3, markerfacecolor='black'))
boxplot.set_xticklabels(boxplot.get_xticklabels(), fontsize=7);
boxplot.set_yticklabels(boxplot.get_yticklabels(), fontsize=7);
plt.tight_layout()
plt.show()
```

```python
# Visualising Categorical variables through boxplots

plt.figure(figsize=(16, 4.5))
plt.subplot(1, 4, 1)
sb.set_palette('tab10')
sb.boxplot(x = 'fueltype', y= 'price', hue='fueltype', data = CarData,
           flierprops=dict(marker='d', markersize=5, markerfacecolor='black'))
plt.subplot(1, 4, 2)
sb.boxplot(x = 'cylindernumber', y= 'price', hue='cylindernumber', data = CarData,
           flierprops=dict(marker='d', markersize=5, markerfacecolor='black'))
plt.subplot(1, 4, 3)
sb.boxplot(x = 'enginelocation', y= 'price', hue='enginelocation', data = CarData,
           flierprops=dict(marker='d', markersize=5, markerfacecolor='black'))
plt.subplot(1, 4, 4)
sb.boxplot(x = 'doornumber', y= 'price', hue='doornumber', data = CarData,
           flierprops=dict(marker='d', markersize=5, markerfacecolor='black'))
plt.tight_layout()
plt.show()
```

```python
#creating a dict 'company price' where key = car's company and value = median of their price

company_price = {}
for company in CarData['CarCompany'].unique():
    company_price[company] = float(CarData[CarData['CarCompany'] == company]['price'].median())

print(company_price)
```

```python
# division in the buckets of low i.e. below 10000, medium i.e. range b/w 10000 and 20000 and high i.e. above 20000

company_price_buckets = {
    company: 'low' if price < 10000 else 'med' if 10000 <= price <= 20000 else 'high'
    for company, price in company_price.items()
}

# mapping the company_price on the 'CarCompany' column in the dataset
CarData['CarCompany'] = CarData['CarCompany'].apply(lambda x: company_price_buckets[x])

# checking the column's unique values

CarData.CarCompany.unique()
```

```python
#creating dummy variable for catwegorical variables
CarData = pd.get_dummies(CarData.drop('CarName', axis=1))

#checking the dataset
CarData.head()
# row_str = CarData.iloc[0].to_string()
# print(row_str)
```

```python
# importing required library
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

# CarData_train, CarData_test = train_test_split(CarData, test_size=0.3, random_state=100)
CarData_train, CarData_test = train_test_split(CarData.drop('car_ID', axis=1), test_size=0.3, random_state=99)

# shape of the train and test sets
print(CarData_train.shape)
print(CarData_test.shape)
```

```python
#importing required function for scaling

from sklearn.preprocessing import MinMaxScaler
```

```python
# CarData_trainScaled = scaler.transform(CarData_train)
# CarData_testScaled = scaler.transform(CarData_test)

X_train = CarData_train.drop('price', axis=1)
y_train = CarData_train['price']

X_test = CarData_test.drop('price', axis=1)
y_test = CarData_test['price']

scaler = MinMaxScaler()
scaler.fit(CarData.drop(['price', 'car_ID'], axis=1))
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# print(X_train)
# print(X_train_scaled)
# print(y_train)
# y_train.head()
# row_str = X_test.iloc[0].to_string()
# print(row_str)
```

```python
# Train your model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train)
# model.fit(CarData.drop('price', axis=1), CarData['price'])
```

```python
# Making predictions
predicted_price = model.predict(X_test_scaled)
```

```python
# Show predictions
predicted_price
```

```python
# Plotting y_test and y_pred to understand the spread.
plot = sb.scatterplot(x=y_test, y=predicted_price)
plot.set_xlabel("y_test")
plot.set_ylabel('y_pred')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, predicted_price)

print(f"R^2 score: {r2}")
```
